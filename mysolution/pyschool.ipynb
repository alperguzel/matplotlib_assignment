```python
import pandas as pd
import os
import numpy as np
```

```python
path_school = os.path.join("raw_data", "schools_complete.csv")
path_student = os.path.join("raw_data", "students_complete.csv")

school_data = pd.read_csv(path_school)
student_data = pd.read_csv(path_student)
student_data.head()
```

```python

school_data.head(15)
```

```python
total_schools = school_data["School ID"].count()
total_schools
```

```python
total_student = student_data['Student ID'].count()
total_student
```

```python
total_budget = school_data["budget"].sum()
total_budget
```

```python
ave_math = student_data['math_score'].mean()
ave_math
```

```python
ave_read = student_data['reading_score'].mean()
ave_read
```

```python
#passing score > 70
math_pass = student_data.loc[student_data['math_score'] > 70]
num_math_pass = math_pass['Student ID'].count()
per_math = num_math_pass / total_student * 100
per_math
```

```python
read_pass = student_data.loc[student_data['reading_score'] > 70]
num_read_pass = read_pass['Student ID'].count()
per_read = num_read_pass / total_student * 100
per_read
```

```python
#overall
overall = (student_data['math_score'] > 70) & (student_data['reading_score'] > 70)
over_all = student_data.loc[overall, :]
both_success =over_all['Student ID'].count()
per_both = both_success / total_student * 100
per_both
```

```python
district_sum = pd.DataFrame({
    'Total Schools': [total_schools],
    'Total Stidents': [total_student],
    'Total Budget': [total_budget],
    'Average Math Score': [ave_math],
    'Average Reading Score': [ave_read],
    '% Passing Math': [per_math],
    '% Passing Reading': [per_read],
    'Overall Passing Rate': [per_both]
})

district_sum
```

```python
new_sch_data = school_data.set_index('name')
new_sch_data
```

```python
#i need just type and budget columns
new_sch_data = new_sch_data.iloc[:, [1,3]]
new_sch_data
```

```python
# well i need to add the other data series
new_sch_data['Per Student Budget'] = new_sch_data['budget'] / sum_stu_of_sch
new_sch_data
```

```python
new_sch_data['Total Students'] = sum_stu_of_sch
new_sch_data['Average Math Score'] = ave_math_of_sch
new_sch_data['Average Reading Score'] = ave_read_of_sch
new_sch_data['% Passing Math Rate'] = rate_of_math_pass
new_sch_data['% Passing Reading Rate'] = rate_of_read_pass
new_sch_data['% Overall Passing Rate'] = rate_of_overall_pass
new_sch_data
```

```python
# so you asked me add $ for be sure i know how to do.
def dollar_covert(money):
    dollar = "$" + str(money)
    return dollar
def rounder(number):
    covert = round(int(number),2)
    return covert
org_new_sch_data = new_sch_data["budget"].map(dollar_covert)
org_new_sch_data = pd.DataFrame(org_new_sch_data)

org_new_sch_data['Per Student Budget'] = new_sch_data["Per Student Budget"].map(dollar_covert)


org_new_sch_data['Average Math Score'] = new_sch_data["Average Math Score"].map(rounder)
org_new_sch_data['Average Reading Score'] = new_sch_data["Average Reading Score"].map(rounder)
org_new_sch_data['% Passing Math Rate'] = new_sch_data["% Passing Math Rate"].map(rounder)
org_new_sch_data['% Passing Reading Rate'] = new_sch_data["% Passing Reading Rate"].map(rounder)
org_new_sch_data['% Overall Passing Rate'] = new_sch_data["% Overall Passing Rate"].map(rounder)

org_new_sch_data
```
